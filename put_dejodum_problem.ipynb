```python
import pandas as pd
lst1 = ['up','up','up','down','down','down','down' ]

lst2 = [10,14,20,50,50,30,40]

lst3 = [10,16,27,50,47,30,40]

lst4 = [10,16,30,50,45,27,40]

df = pd.DataFrame(

    {'direction': lst1,

     'price_t1': lst2,

     'price_t2': lst3,

     'price_t3': lst4

    })
```

```python
df
```

```python
df_output = {'direction':[],'output':[]}
```

```python
for i ,row in df.iterrows():
    if row.direction=='up':
        if(row.price_t1==row.price_t2 and row.price_t2 == row.price_t3):
            df_output['direction'].append('up')
            df_output['output'].append('price_t1')
        elif(row.price_t1>row.price_t2 and row.price_t1>row.price_t3):
            df_output['direction'].append('up')
            df_output['output'].append('price_t1')
        elif(row.price_t2>row.price_t1 and row.price_t2>row.price_t3):
            df_output['direction'].append('up')
            df_output['output'].append('price_t2')
        elif(row.price_t3>row.price_t1 and row.price_t3>row.price_t2):
            df_output['direction'].append('up')
            df_output['output'].append('price_t3')
        elif(row.price_t1>row.price_t2 and row.price_t1==row.price_t3):
            df_output['direction'].append('up')
            df_output['output'].append('price_t1')
          
        elif(row.price_t1>row.price_t3 and row.price_t1==row.price_t2):
            df_output['direction'].append('up')
            df_output['output'].append('price_t1')
            
        elif(row.price_t2>row.price_t1 and row.price_t2==row.price_t3):
            df_output['direction'].append('up')
            df_output['output'].append('price_t2')
            
        elif(row.price_t2>row.price_t3 and row.price_t2==row.price_t1):
            df_output['direction'].append('up')
            df_output['output'].append('price_t1')
            
        elif(row.price_t3>row.price_t2 and row.price_t3==row.price_t1):
            df_output['direction'].append('up')
            df_output['output'].append('price_t1')
        elif(row.price_t3>row.price_t1 and row.price_t3==row.price_t2):
            df_output['direction'].append('up')
            df_output['output'].append('price_t2')
    if row.direction=='down':
        if(row.price_t1==row.price_t2 and row.price_t2 == row.price_t3):
            df_output['direction'].append('down')
            df_output['output'].append('price_t1')
        elif(row.price_t1<row.price_t2 and row.price_t1<row.price_t3):
            df_output['direction'].append('down')
            df_output['output'].append('price_t1')
        elif(row.price_t2<row.price_t1 and row.price_t2<row.price_t3):
            df_output['direction'].append('down')
            df_output['output'].append('price_t2')
        elif(row.price_t3<row.price_t1 and row.price_t3<row.price_t2):
            df_output['direction'].append('down')
            df_output['output'].append('price_t3')
        elif(row.price_t1<row.price_t2 and row.price_t1==row.price_t3):
            df_output['direction'].append('down')
            df_output['output'].append('price_t1')
          
        elif(row.price_t1<row.price_t3 and row.price_t1==row.price_t2):
            df_output['direction'].append('down')
            df_output['output'].append('price_t1')
            
        elif(row.price_t2<row.price_t1 and row.price_t2==row.price_t3):
            df_output['direction'].append('down')
            df_output['output'].append('price_t2')
            
        elif(row.price_t2<row.price_t3 and row.price_t2==row.price_t1):
            df_output['direction'].append('down')
            df_output['output'].append('price_t1')
            
        elif(row.price_t3<row.price_t2 and row.price_t3==row.price_t1):
            df_output['direction'].append('down')
            df_output['output'].append('price_t1')
        elif(row.price_t3<row.price_t1 and row.price_t3==row.price_t2):
            df_output['direction'].append('down')
            df_output['output'].append('price_t2')
                            
```
